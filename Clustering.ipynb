```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
df = pd.read_csv("study_d.csv") 
df.head()
```

```
new_df = df.drop(['Name'], axis=1)
final_df = new_df.drop(['Type'], axis=1)
```

```
final_df.head()
```

```
kmeans = KMeans(n_clusters=5).fit(final_df)
centroids = kmeans.cluster_centers_
print(centroids)
figure1 = plt.figure(figsize=(10,9), dpi=100)
plt.scatter(final_df['Lat'], final_df['Long'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=150 , marker='*', label ='centroid')
```
